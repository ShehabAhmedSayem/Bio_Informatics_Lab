```python
def peptide_mass_dict_generate():
    peptideMass = {"G":57,"A":71,"S":87,"P":97,"V":99,"T":101,
                   "C":103,"I":113,"L":113,"N":114,"D":115,"K":128,
                   "Q":128,"E":129,"M":131,"H":137,"F":147,"R":156,"Y":163,"W":186}
    return peptideMass

def generate_theoretical_spectrum(s):
    peptideMass = peptide_mass_dict_generate()

    l = len(s)
    s1 = s+s

    ans = [0]
    for i in range(1,l):
        for j in range(l):
            sub = s1[j:j+i]
            n=0
            for k in sub:
                n += peptideMass[k]
            ans.append(n)
    n=0
    for k in s:
        n += peptideMass[k]
    ans.append(n)
    ans.sort()
    
    return ans

def generate_linear_spectrum(s):
    peptideMass = peptide_mass_dict_generate()

    l = len(s)

    ans = [0]
    for i in range(1,l):
        for j in range(l-i+1):
            sub = s[j:j+i]
            n=0
            for k in sub:
                n += peptideMass[k]
            ans.append(n)
    n=0
    for k in s:
        n += peptideMass[k]
    ans.append(n)
    ans.sort()
    
    return ans


def expand(peptide_list): 
    new_peptide_list = [] 
    single_peptide_list = ["G", "A", "S", "P", "V", "T", "C", "I", 
                          "L", "N", "D", "K", "Q", "E", "M", "H", 
                          "F", "R", "Y", "W"] 
    for p in peptide_list: 
        for i in single_peptide_list: 
            new_peptide_list.append(p+i)
            
    return new_peptide_list 


def mass(peptide):
    m = 0
    peptideMass = peptide_mass_dict_generate()
    for p in peptide:
        m+=peptideMass[p]
    return m


def match(theory_spec, experimental_spectrum):
    t_spec = theory_spec.copy()
    exp_spec = experimental_spectrum.copy()
    ans = False
    
    for i in theory_spec:
        if(i in exp_spec):
            t_spec.remove(i)
            exp_spec.remove(i)
    if len(t_spec)==0 and len(exp_spec)==0:
        ans = True
    
    return ans


def consistent(pep, spec):
    linear_spec = generate_linear_spectrum(pep)
    l_spec = linear_spec.copy()
    exp_spec = spec.copy()
    ans = False
    for i in linear_spec:
        if(i in exp_spec):
            l_spec.remove(i)
    
    if len(l_spec)==0:
        ans = True
    
    return ans

def cycloPeptideSeq(spec):
    parentmass = max(spec)
    output_list = []
    peptide_list = [''] 
    
    while peptide_list: 
        
        peptide_list = expand(peptide_list)
        new_peptide_list = peptide_list.copy()
        
        for peptide in peptide_list:
            
            if mass(peptide) == parentmass:
                theory_spec = generate_theoretical_spectrum(peptide)
                if match(theory_spec, spec):
                    output_list.append(peptide)
                new_peptide_list.remove(peptide)
            
            elif not consistent(peptide, spec):
                new_peptide_list.remove(peptide)
                
            elif mass(peptide) > parentmass:
                new_peptide_list.remove(peptide)
        
        peptide_list = new_peptide_list.copy()
        
    return output_list
    
#ans = cycloPeptideSeq([0, 97, 99, 99, 99, 113, 129, 129, 131, 196, 212, 228, 228, 228, 228, 230, 
#                       242, 325, 327, 327, 327, 341, 341, 341, 359, 426, 438, 440, 440, 456, 456, 
#                       458, 470, 537, 555, 555, 555, 569, 569, 569, 571, 654, 666, 668, 668, 668,
#                       668, 684, 700, 765, 767, 767, 783, 797, 797, 797, 799, 896])
#cycloPeptideSeq([0, 114, 129, 243])

ans = cycloPeptideSeq([0, 113, 128, 186, 241, 299, 314, 427])

peptideMass = peptide_mass_dict_generate()
ans_list = []
for i in ans:
    s=''
    k=0
    for j in i:
        if k!=0:
            s += '-'+str(peptideMass[j])
        else:
            k=1
            s += str(peptideMass[j])
    ans_list.append(s)
ans_list = sorted(set(ans_list))[::-1]
ans_s = ''
for s in ans_list:
    ans_s += s+' '
print(ans_s)
```
